```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from itertools import chain
```

```python
file_url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQfwFLOECjiMWuBd6Zu1GCpQIPJiRd73w-0rVAYZE4XfvBwmKdcN6JfLRdJP63kMUUeeHRD806uQe0z/pub?output=csv'
df_weather = pd.read_csv(file_url)
df_weather['Date'] = pd.to_datetime(df_weather['Date'])
```

```python
df_scale = df_weather.copy(deep=True)
count=0
for col in df_scale:
  if count>3:
    df_scale[col] = ((df_scale[col]-df_scale[col].min())/(df_scale[col].max()-df_scale[col].min()))
  else:
    count +=1
```

```python
df_weather["Longitude"] = df_weather["Longitude"].round(2)
df_weather["Latitude"] = df_weather["Latitude"].round(2)
df_jan = df_weather[(df_weather.Date.dt.month == 1)]
key_list = list(df_scale.groupby(["Longitude","Latitude"]).groups.keys())
```

# **Linear Regression**

we can create the chart for each month in every location

```python
lin_month = 4
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
sum_error=0

for key in key_list:
  df_loc = df_scale[(df_scale['Longitude'] == key[0]) & (df_scale['Latitude'] == key[1]) & (df_scale.Date.dt.month == lin_month)][['Date', 'Max Temperature', 'Min Temperature', "Longitude", "Latitude", "Elevation"]]
  df_avg = df_loc.groupby(df_loc.Date.dt.year)[['Max Temperature','Min Temperature']].apply(sum)/df_loc.Date.dt.day.max()
  df_avg.reset_index(inplace = True)

  train_data = df_avg[df_avg["Date"]<2005]
  valid_data = df_avg[df_avg["Date"]>2004]
  print("Longitude: ",df_loc["Longitude"].head(1).values[0])
  print("Latitude: ", df_loc["Latitude"].head(1).values[0])
  print("Elevation: ",df_loc["Elevation"].head(1).values[0])
  reg = LinearRegression().fit(np.array(train_data["Date"]).reshape(-1,1), np.array(train_data["Max Temperature"]))

  y_pred = reg.predict(np.array(valid_data["Date"]).reshape(-1,1))
  plt.figure(figsize=(10, 6))
  plt.scatter(df_avg["Date"], np.array(df_avg["Max Temperature"]))
  plt.plot(np.array(valid_data["Date"]),y_pred)
  plt.scatter(train_data["Date"], np.array(train_data["Max Temperature"]))
  print("MSE Location",count+1,mean_squared_error(valid_data["Max Temperature"], y_pred, squared=False))
  sum_error += mean_squared_error(valid_data["Max Temperature"], y_pred, squared=False)
  y_test = np.array(range(2015,2030), dtype=np.int64)
  y_pred_future = reg.predict(y_test.reshape(-1,1))
  plt.plot(np.array(y_test), y_pred_future,color = "b")
  plt.ylim([0.0, 1.0])
  plt.show()
  print()
sum_error
```

# **Polynomial Regression**

```python
pol_month = 4
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
select_dict={}
sum_error=0.0
for key in key_list:
  df_loc = df_scale[(df_scale['Longitude'] == key[0]) & (df_scale['Latitude'] == key[1]) & (df_scale.Date.dt.month == pol_month)][['Date', 'Max Temperature', 'Min Temperature', "Longitude", "Latitude", "Elevation"]]
  df_avg = df_loc.groupby(df_loc.Date.dt.year)[['Max Temperature','Min Temperature']].apply(sum)/df_loc.Date.dt.day.max()
  df_avg.reset_index(inplace = True)
  train_data = df_avg[df_avg["Date"]<2005]
  valid_data = df_avg[df_avg["Date"]>2004]
  error_dict = {2:[],3:[],4:[],5:[],6:[],7:[],8:[],9:[],10:[]}
  
  for i in range(2,11):
    poly = PolynomialFeatures(degree = i)
    poly_features = poly.fit_transform(np.array(train_data["Date"]).reshape(-1,1))
    poly_reg_model = LinearRegression()
    poly_reg_model.fit(poly_features, train_data["Max Temperature"])
    valid_feature = poly.fit_transform(np.array(valid_data["Date"]).reshape(-1,1))
    y_predicted = poly_reg_model.predict(valid_feature)
    error_dict[i].append(mean_squared_error(valid_data["Max Temperature"], y_predicted, squared=False))

  print("Longitude: ",df_loc["Longitude"].head(1).values[0])
  print("Latitude: ", df_loc["Latitude"].head(1).values[0])
  print("Elevation: ",df_loc["Elevation"].head(1).values[0])
  sort_dict = sorted(error_dict.items(), key=lambda x:x[1])
  print(sort_dict)
  print("Degree ", sort_dict[0][0] ,":",sort_dict[0][1])
  if(sort_dict[0][0] in select_dict):
      select_dict[sort_dict[0][0]] = select_dict[sort_dict[0][0]]+1
  else:
      select_dict[sort_dict[0][0]] = 1
  error_dict
  plt.figure(figsize=(10, 6))
  plt.plot(list(error_dict.keys()), list(error_dict.values()))
  sum_error += sort_dict[0][1][0]

  plt.figure(figsize=(10, 6))
  plt.scatter(valid_data["Date"],valid_data["Max Temperature"])
  plt.plot(valid_data["Date"], y_predicted)
  y_test = np.array(range(2015,2030), dtype=np.int64)
  future_feature = poly.fit_transform(y_test.reshape(-1,1))
  y_future = poly_reg_model.predict(future_feature)
  plt.plot(y_test, y_future)
  plt.ylim([0.0, 1.0])
  plt.show()
  print()
print("What degree this modle chose and how many times chose ", select_dict)
print("Sum of error: ", sum_error)
```

```python
import statistics
variance_arr = []
valid_sum = 0
for month in range(1,13):
  for key in key_list:
    df_loc = df_scale[(df_scale['Longitude'] == key[0]) & (df_scale['Latitude'] == key[1]) & (df_scale.Date.dt.month == month)][['Date', 'Max Temperature', 'Min Temperature', "Longitude", "Latitude", "Elevation"]]
    df_avg = df_loc.groupby(df_loc.Date.dt.year)[['Max Temperature','Min Temperature']].apply(sum)/df_loc.Date.dt.day.max()
    df_avg.reset_index(inplace = True)
    valid_data = df_avg[df_avg["Date"]>2004]
    valid_sum += statistics.stdev(valid_data["Max Temperature"])
  variance_arr.append(valid_sum/35)
  valid_sum = 0
```

```python
select_dict={}
err_sum=0.0
sumerror = []
for month in range(1,13):
  for key in key_list:
    df_loc = df_scale[(df_scale['Longitude'] == key[0]) & (df_scale['Latitude'] == key[1]) & (df_scale.Date.dt.month == month)][['Date', 'Max Temperature', 'Min Temperature', "Longitude", "Latitude", "Elevation"]]
    df_avg = df_loc.groupby(df_loc.Date.dt.year)[['Max Temperature','Min Temperature']].apply(sum)/df_loc.Date.dt.day.max()
    df_avg.reset_index(inplace = True)
    df_avg.reset_index(inplace = True)
    train_data = df_avg[df_avg["Date"]<2005]
    valid_data = df_avg[df_avg["Date"]>2004]
    error_dict = {2:[],3:[],4:[],5:[],6:[],7:[],8:[],9:[],10:[]}
    
    for i in range(2,11):
      poly = PolynomialFeatures(degree = i)
      poly_features = poly.fit_transform(np.array(train_data["Date"]).reshape(-1,1))
      poly_reg_model = LinearRegression()
      poly_reg_model.fit(poly_features, train_data["Max Temperature"])
      valid_feature = poly.fit_transform(np.array(valid_data["Date"]).reshape(-1,1))
      y_predicted = poly_reg_model.predict(valid_feature)
      error_dict[i].append(mean_squared_error(valid_data["Max Temperature"], y_predicted, squared=False))

    sort_dict = sorted(error_dict.items(), key=lambda x:x[1])
    if(sort_dict[0][0] in select_dict):
        select_dict[sort_dict[0][0]] = select_dict[sort_dict[0][0]]+1
    else:
        select_dict[sort_dict[0][0]] = 1
    error_dict
    err_sum += sort_dict[0][1][0]
  sumerror.append(err_sum)
  err_sum = 0
```

```python
sumerror_lin = []
sum_lin=0
for month in range(1,13):
  for key in key_list:
    df_loc = df_scale[(df_scale['Longitude'] == key[0]) & (df_scale['Latitude'] == key[1]) & (df_scale.Date.dt.month == month)][['Date', 'Max Temperature', 'Min Temperature', "Longitude", "Latitude"]]
    df_avg = df_loc.groupby(df_loc.Date.dt.year)[['Max Temperature','Min Temperature']].apply(sum)/df_loc.Date.dt.day.max()
    df_avg.reset_index(inplace = True)
    df_avg.reset_index(inplace = True)

    train_data = df_avg[df_avg["Date"]<2005]
    valid_data = df_avg[df_avg["Date"]>2004]
    reg = LinearRegression().fit(np.array(train_data["Date"]).reshape(-1,1), np.array(train_data["Max Temperature"]))

    y_pred = reg.predict(np.array(valid_data["Date"]).reshape(-1,1))
    sum_lin += mean_squared_error(valid_data["Max Temperature"], y_pred, squared=False)
    y_test = np.array(range(2015,2030), dtype=np.int64)
    y_pred_future = reg.predict(y_test.reshape(-1,1))
    count += 1
  sumerror_lin.append(sum_lin)
  sum_lin=0
```
